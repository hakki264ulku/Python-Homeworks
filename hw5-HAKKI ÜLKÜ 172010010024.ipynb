```python
#import the libaries
print("The function is 'y= cos(2*pi*(a+x/20))' \n\n")
from math import cos, pi
import matplotlib.pyplot as plt
%matplotlib inline

y1=[]#empty list
y2=[]#empty list
y3=[]#empty list

n=int(input("Enter the range of x values between 0-40 : "))#user will enter the range number

for a in range(0,3):#this loop is gets a=0,1,2
    for x in range(n):#then get all the range between 0-user input(n)
        if a==0:#check if a=0 
            f=cos(2*pi*(x/20)+a)#then do this calcualtion
            y1.append(f)#add to y1 list
        elif a==1:#again check if a=1
            f=cos(2*pi*(x/20)+a)#then do this calcualtion
            y2.append(f)#add to y2 list
        else:#otherwise if a=2
            f=cos(2*pi*(x/20)+a)#then do this calcualtion
            y3.append(f)#add to y3 list
x=[i for i in range(n)]#this variable stores numbers between 0 to n as x axis 

plt.plot(x,y1,'o-',color='blue')#plot x axis and y1 
plt.plot(x,y2,'D-',color='green')#plot x axis and y2
plt.plot(x,y3,'>--',color='red')#plot x axis and y3
plt.title("Function graph")#title
plt.legend(['a=0','a=1','a=2'])
plt.xlabel("x")#x label
plt.ylabel("$y_a$")#y label
plt.grid(True)#the grid
plt.show()#lastly show the plot 
            
```
